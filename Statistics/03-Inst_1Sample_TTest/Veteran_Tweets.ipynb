```python
# Dependencies
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
```

```python
# Generate "Simulated" Data
population_sentiment = stats.gennorm.rvs(beta=25, size=1000)

veteran_sentiment1 = stats.poisson.rvs(loc=-40, mu=10, size=200) / 100
veteran_sentiment2 = stats.poisson.rvs(loc=-10, mu=15, size=100) / 100
veteran_sentiment3 = stats.poisson.rvs(loc=10, mu=15, size=100) / 100
veteran_sentiment = np.concatenate((veteran_sentiment1, veteran_sentiment2, veteran_sentiment3))

print(population_sentiment.mean())
print(veteran_sentiment.mean())
```

```python
# Plot the Data
plt.scatter(np.arange(0, len(population_sentiment)), population_sentiment, c="r", alpha=0.09, label="General")
plt.scatter(np.arange(0, len(veteran_sentiment)), veteran_sentiment,  c="b", label="Veterans")
plt.legend(loc="best")
plt.xlabel("Tweet #")
plt.ylabel("Sentiment")
plt.title("Sentiment of Veterans vs. General Population")
plt.show()
```

```python
# Run 1-Sample T-Test
stats.ttest_1samp(a=veteran_sentiment, popmean=population_sentiment.mean())
```
